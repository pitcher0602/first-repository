```python
import random

class Account:
    count = 0

    def __init__(self, name, balance):
        self.bank = "SC은행"
        self.name = name
        self.balance = balance
        
        first = random.randint(100,999)
        second = random.randint (10,99)
        third = random.randint(100000,999999)
        self.account_number = str(first)+"-"+str(second) + "-" +str(third)
        
        self.deposit_count = 0
        self.deposit_history = []
        self.withdraw_history = []

        Account.count += 1

    def get_account_num():
        return Account.count

    def deposit(self, money):
        if money < 1:
            return

        self.balance += money
        self.deposit_count += 1
        self.deposit_history.append(money)

        if self.deposit_count % 5 == 0:
            self.balance += self.balance * 0.01

    def withdraw(self, money):
        if money > self.balance:
            return

        self.balance -= money
        self.withdraw_history.append(money)

    def display_info(self):
        print("(은행이름:", self.bank,
              ", 예금주:", self.name,
              ", 계좌번호:", self.account_number,
              ", 잔고:", format(self.balance, ","), "원)")

    def deposit_history_print(self):
        print("입금 내역:", self.deposit_history)

    def withdraw_history_print(self):
        print("출금 내역:", self.withdraw_history)
```

```python

a = Account("권지훈", 10000)

a.deposit(1000)
a.deposit(1000)
a.deposit(1000)
a.deposit(2000)
a.deposit(1000)   

a.withdraw(5000)

a.display_info()
a.deposit_history_print()
a.withdraw_history_print()
```

```python
import random

class Character:
    def __init__ (self, name,level,hp,attack,defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.hp>0

    def take_damage(self,damage):
        my_damage = damage - self.defense
        if my_damage <= 0:
            my_damage = 0
        self.hp -= my_damage

    def attack_target(self,target):
        damage = self.attack
        target.take_damage(damage)

class Player(Character):
    def __init__(self,name):
        Character.__init__(self,name,1,100,20,5)
        self.exp = 0

    def gain_experience(self):
        self.exp +=10
    
    def level_up(self):
        if self.exp >=50:
            self.level +=1
            self.exp -= 50
            self.attack +=10
            self.defense +=5
        

class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        Character.__init__(self, name, level, hp, attack, defense)

def battle(player, monster):
    print("전투 시작")
    print(player.name, "vs",monster.name)

    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        print(monster.name,"체력",monster.hp) 

        if not monster.is_alive():
            print("전투승리")
            player.exp += monster.level* 20
            player.level_up()
            return
        monster.attack_target(player)
        print(player.name,"체력:",player.hp)

        if not player.is_alive():
            print("전투패배..")
            return
        
def main():
    Monster_dict = {
        "슬라임" :1,
        "고블린" :2,
        "오크":3 } 
    name =input("")
    player = Player(name)

    for monster_name in Monster_dict:
        level = Monster_dict[monster_name]
        monster = Monster(monster_name,level)

        battle(player,monster)

        if not player. is_alive():
            print("게임오버")
            return
        print("우왕")
```

```python
main()
```
